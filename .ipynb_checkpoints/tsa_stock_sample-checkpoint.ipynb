### The first obvious thing to note, aside from the dips in the tail end corresponding to the market crashes in between 2018 and 2019 , is that the data is clearly non-stationary.


### Check the staionarity of the data
A TS is said to be stationary if its statistical properties such as mean, variance remain constant over time. But why is it important? Most of the TS models work on the assumption that the TS is stationary. Intuitively, we can sat that if a TS has a particular behaviour over time, there is a very high probability that it will follow the same in the future. Also, the theories related to stationary series are more mature and easier to implement as compared to non-stationary series.

Stationarity is defined using very strict criterion. However, for practical purposes we can assume the series to be stationary if it has constant statistical properties over time, ie. the following:

constant mean,
constant variance,
an autocovariance that does not depend on time.

## Plotting Rolling Statistics:
We can plot the moving average or moving variance and see if it varies with time. By moving average/variance I mean that at any instant ‘t’, we’ll take the average/variance of the last year, i.e. last 12 months. But again this is more of a visual technique.


```python
def test_stationarity(timeseries):
    
    #Determing rolling statistics
#     ts_log.rolling(12).mean()
    #rolmean = pd.rolling_mean(timeseries, window=12)
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()

    #Plot rolling statistics:
    plt.figure(figsize=(20,10))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    #print('Results of Dickey-Fuller Test:')
    #dftest = adfuller(timeseries, autolag='AIC')
    #dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    #for key,value in dftest[4].items():
    #    dfoutput['Critical Value (%s)'%key] = value
    #print(dfoutput)

test_stationarity(amzn_data)
```

####  As we can see in the above plot the rolling mean and rolling stddev is not constant over a period of time. The data is non stationary.This makes sense for market data as it tends to go up in the long run more than it goes down. This is a problem for time series analysis though as non-stationary data is hard to reason about. 

#### The first thing we can try is a first difference of the series. In other words, subtract the previous value t-1 from the current value t to get the difference d(t)

```python
amz_df = amzn_data.copy()
amz_df['First Difference'] = amz_df['Close'] - amz_df['Close'].shift()
amz_df['First Difference'].plot(figsize=(16,6))
```

#### Look at the graph. It’s very small early on and steadily increases over time. This is a sign that the data is not only non-stationary but also exponentially increasing. The magnitude of the day-to-day variations at present day completely dwarf the magnitude of the changes in previous years. To deal with this, we have t normalize the data we’ll apply a log transform to the original series (data before first difference)

#### log tranformation is done to flatten the data i.e from exponential curve to a linear curve 

```python
#applying log transform
amzn_data['Log'] = amzn_data['Close'].apply(lambda x: np.log(x))  
amzn_data['Log'].plot(figsize=(16, 6))
```

#### In the above graph the original closing price after applying the log transform applied to “flatten” the data from an exponential curve to a linear curve.
#### Since we can see the data is still non stationary, we can apply log transformation on the variance
#### One way to visually see the effect that the log transform had is to analyze the variance over time. We can use a rolling variance statistic and compare both the original series and the logged series.

```python
amzn_data['Original Variance'] = amzn_data['Close'].rolling(10, min_periods=None,  center=True).var()

amzn_data['Log Variance'] = amzn_data['Log'].rolling(10, min_periods=None, center=True).var()
#print(amzn_data['Original Variance'])
#print(amzn_data['Log Variance'])

fig, ax = plt.subplots(2, 1, figsize=(13, 12))  
amzn_data['Original Variance'].plot(ax=ax[0], title='Original Variance')  
amzn_data['Log Variance'].plot(ax=ax[1], title='Log Variance')  
fig.tight_layout()
```

#### From this view, it’s clear that our transformation has made the variance relatively constant.
#### To see the better result we will do the first difference of the logged values.

```python
amzn_data['logged first diff'] = amzn_data['Log'] - amzn_data['Log'].shift()
amzn_data['logged first diff'].plot(figsize=(16, 12))
```

#### We can see now we have a stationary time series model of daily changes to the close prices of the stock. 
#### Now let’s create some lag variables y(t-1), y(t-2) etc. and examine their relationship to y(t). We’ll look at 1 and 2-day lags along with weekly and monthly lags to look for “seasonal” effects

```python
amzn_data['Lag 1']  = amzn_data['logged first diff'].shift(1)
amzn_data['Lag 2']  = amzn_data['logged first diff'].shift(2)
amzn_data['Lag W']  = amzn_data['logged first diff'].shift(5)
amzn_data['Lag M']  = amzn_data['logged first diff'].shift(30)


sns.scatterplot(x=amzn_data['Close'],y=amzn_data['Lag 1'])
#sns.scatterplot(x=amzn_data['Close'],y=amzn_data['Lag 2'])
#sns.scatterplot(x=amzn_data['Close'],y=amzn_data['Lag W'])
#sns.scatterplot(x=amzn_data['Close'],y=amzn_data['Lag M'])
```

```python
sns.jointplot('logged first diff', 'Lag 1', amzn_data, kind='reg', size=13)
```

Notice how tightly packed the mass is around 0. It also appears to be pretty evenly distributed — the marginal distributions on both axes are roughly normal. This seems to indicate that knowing the index value one day doesn’t tell us much about what it will do the next day.

To find the correlation among other lag values we will use a statsmodel library to detect the aotu corelation in data 

```python

lag_correlations = acf(amzn_data['logged first diff'].iloc[1:] , nlags=100,fft=None)
lag_partial_correlations = pacf(amzn_data['logged first diff'].iloc[1:],nlags=100)

fig, ax = plt.subplots(figsize=(16,6))  
ax.plot(lag_correlations, marker='o', linestyle='--')

ax.plot(lag_partial_correlations, marker='o', linestyle='--')

```

The auto-correlation and partial-autocorrelation results are very close to each other. What this shows is that there is no significant (> 0.2) correlation between the value at time t and at any time prior to t up to 500 steps behind. In order words, the series is a random walk.
### Decomposition
Another interesting technique we can try is a decomposition. This is a technique that attempts to break down a time series into trend, seasonal, and residual factors. Statsmodels comes with a decompose function out of the box.


```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(amzn_data['Log'], model='additive', freq=30)
#print(decomposition)
fig = plt.figure()  
fig = decomposition.plot()
```
