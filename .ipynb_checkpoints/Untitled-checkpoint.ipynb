```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima_model import ARIMA
import matplotlib.pylab as plt
import datetime

df = pd.read_csv('stock_data.csv',index_col=False)
cond = {'1' : 'AMZN', '2': 'TIF','3':'HD','4':"AAPL",'5':"WMT"}

data_1 = df.loc[df.ticker == cond['1'], ['Date','Close']]  #for AMZN
#print(data_1)

avg= data_1['Close']
avg=list(avg)
res = pd.Series(avg, index=pd.to_datetime(data_1['Date'],format='%Y-%m-%d'))

ts=np.log(res)

ts_diff = ts - ts.shift()
ts_diff.dropna(inplace=True)

r = ARIMA(ts,(0,1,0))
r = r.fit(disp=-1)

dates = pd.date_range('2019-09-21','2019-09-25')

pred = r.predict(start=1258,end=1265)


predictions_ARIMA_diff = pd.Series(pred, copy=True)
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
predictions_ARIMA_log = pd.Series(ts)

predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value=0)

print(predictions_ARIMA_log)
predictions_ARIMA = np.exp(predictions_ARIMA_log)

#print(predictions_ARIMA)


#
plt.plot(ts_diff)
plt.plot(predictions_ARIMA[2:])

#plt.show()

```
